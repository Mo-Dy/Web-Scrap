```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from scipy import stats
from datetime import datetime
from sklearn import preprocessing
```

```python
model = LinearRegression(fit_intercept=True)
model
```

```python
X = pd.DataFrame(fin_data['Prices'])
y = pd.DataFrame(fin_data['Before_Sall'])
scores = []
kfold = KFold(n_splits=3, shuffle=True, random_state=42)
for i, (train, test) in enumerate(kfold.split(X, y)):
 model.fit(X.iloc[train,:], y.iloc[train,:])
 score = model.score(X.iloc[test,:], y.iloc[test,:])
 scores.append(score)
print(scores)
```

```python
X['Prices'] = X.index

add_var = pd.get_dummies(X['Prices'], prefix='Prices', drop_first=True)

X = X.join(add_var)

X.drop(columns=['Prices'], inplace=True)
X.head()
```

```python
y['Before_Sall'] = y.index

add_var_y = pd.get_dummies(y['Before_Sall'], prefix='Before_Sall', drop_first=True)

y = y.join(add_var_y)

y.drop(columns=['Before_Sall'], inplace=True)
y.head()
```

```python
price_avg = fin_data.groupby('Type')['Prices'].mean()
price_avg.head()
```

```python
price_count = fin_data.groupby('Type')['Prices'].count().sort_values(ascending=False)
price_count.head()
```

```python
price_mean_count = pd.DataFrame(price_avg)
price_mean_count['price_count']=price_count
price_mean_count.head()
```

```python
sns.set()
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4), dpi = 120)
ax1.hist(price_mean_count['price_count'],bins=50, range=(10,40))
ax2.hist(price_mean_count['Prices'],bins=50, range=(100,1000))
ax1.set_ylabel('Number of Type')
ax1.set_xlabel('Price counts')
ax2.set_ylabel('Number of Type')
ax2.set_xlabel('Price values')
plt.show()
```

```python
shoes_price = fin_data.pivot_table(index='ID', columns='Type', values='Prices')
shoes_price.head()
```

```python
forrest_shoes_price = shoes_price['Flat']
forrest_shoes_price.head(10)
```

```python
shoes_like_forest_price = pd.DataFrame(shoes_price.corrwith(forrest_shoes_price), columns=['Commanded'])
shoes_like_forest_price.head(10)
```

```python
shoes_like_forest_price.dropna(inplace=True)
shoes_like_forest_price.head()
```

```python
shoes_like_forest_price.sort_values('Commanded', ascending=False).head()
```

```python
shoes_like_forest_price = shoes_like_forest_price.join(price_mean_count['price_count'])
shoes_like_forest_price.head()
```

```python
recommended_shoes = shoes_like_forest_price[shoes_like_forest_price['price_count']>30].sort_values('Commanded', ascending=False)
recommended_shoes.head()
```
