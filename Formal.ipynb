```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = 'https://www.newchic.com/formal-shoes-c-3617/?country=63&conversionType=US&from=nav'
html = urlopen(url)
```

```python
soup = BeautifulSoup(html)
```

```python
data_name = soup.findAll('p', class_='wom_lst_name ellipsis')
data_name
```

```python
## Clean Data ##
import re
Names = []
for li in data_name:
    str_name = str(li)
    pattern = '<.*?>'
    clean_name = re.sub(pattern, '', str_name)
    Names.append(clean_name)
Names
```

```python
x = pd.Index(range(35,70))
print(x)
```

```python
df = pd.DataFrame(Names,index=x)
df.head()
```

```python
df[0] = df[0].str.strip('\n')
df.head()
```

```python
df.rename(columns={0:'Names'}, inplace=True)
```

```python
df['Names'].replace('\n','', regex=True, inplace=True)
df
```

```python
data_price = soup.findAll('span', class_='price product-price-js')
data_price
```

```python
## Clean Data ##
import re
Prices = []
for li in data_price:
    str_name = str(li)
    pattern = '<.*?>'
    clean_name = re.sub(pattern, '', str_name)
    Prices.append(clean_name)
Prices
```

```python
df1 = pd.DataFrame(Prices, index=x)
df1
```

```python
df1.rename(columns={0:'Prices'}, inplace=True)
df1.head()
```

```python
df1['Prices'].replace('E£','', regex=True, inplace=True)
df1
```

```python
df1['Prices'].replace(',','',regex=True, inplace=True)
df1.tail()
```

```python
df1['Prices'] = pd.to_numeric(df1['Prices'])
df1.tail()
```

```python
df_formal = df.join(df1)
df_formal
```
